```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
import numpy as np
import time

#specify driver path
DRIVER_PATH = r'C:/Users/Jeremy/Desktop/Zurich Project/chromedriver.exe'
driver = webdriver.Chrome(executable_path = DRIVER_PATH)

#Navigate the page
driver.get('https://www.jobstreet.co.id/')

time.sleep(10)
#Input Job 
search_work = driver.find_element_by_id("searchKeywordsField")
search_work.send_keys("Data Analyst")

# #Input Location
# search_loc = driver.find_element_by_id("locationAutoSuggest")
# search_loc.send_keys("Jakarta")

#Search Button
search_but = driver.find_element_by_xpath('//div/button[@data-automation="searchSubmitButton"]')
search_but.click()

#let the driver wait 3 seconds to locate the element before exiting out
driver.implicitly_wait(3) 

titles=[] #
companies=[] #
locations=[] #
links =[] #
salaries = [] # 
times=[] #
descriptions=[] #
comp_profs=[] #
level=[] #
qualifications=[] #
experiences=[] #
types=[] #
specialization=[] #
company_scales=[] 
processed=[]
benefits=[]


for i in range(0,40): #set the max number as you like (used for accessing the number of pages)
    
    job_card = driver.find_elements_by_xpath('.//article[contains(@data-automation,"job-card")]')
    
    for job in job_card:
    
        # Get Job Title
        try:
            title = job.find_element_by_xpath('.//h1[contains(@class,"FYwKg _2Bz3E")]/a/div[contains(@class,"FYwKg _2j8fZ")]').text
        except:
            title = "None"
        titles.append(title)
        
        #Get Company
        try:
            company = job.find_element_by_xpath('.//div[@class="FYwKg"]/div[2]/span[contains(@class,"FYwKg")]').text
        except:
            company = "None"
        companies.append(company)
            
        #Get Salary 
        try:
            salary = job.find_element_by_xpath('.//div[@class="FYwKg"]/span[2][contains(@class,"FYwKg")]').text
        except:
            salary = "-1"
        salaries.append(salary)

        #Get Location
        try:
            location = job.find_element_by_xpath('.//div[@class="FYwKg"]/span[1][contains(@class,"FYwKg")]').text
        except:
            location = "None"
        locations.append(location)

        #Get published time 
        try:
            time = job.find_element_by_xpath('.//time[@class="FYwKg _2Bz3E"]').get_attribute(name="datetime")
        except:
            time = 'None'
        times.append(time)

        #Get redirect link
        try:
            link = job.find_element_by_xpath('.//h1[contains(@class,"FYwKg _2Bz3E")]/a').get_attribute(name="href")
        except:
            link = "None"
        links.append(link)
        
    # Access next page
    try:
        next_page_btn = driver.find_element_by_xpath('.//div[@data-automation="pagination"]/a[2]')
        next_page_btn.click()
    except:
        next_page_btn = driver.find_element_by_xpath('.//div[@data-automation="pagination"]/a[1]') #for the 1st page case
        next_page_btn.click()
    print("Page: {}".format(str(i+2)))
    

# access other information in the main page of each job card
for link in links:
    driver.get(link)
    
    #Get the desc of job (anything in the desc box)
    try:
        desc = driver.find_element_by_xpath('.//div[@data-automation="jobDescription"]').text
    except:
        desc = "None"
    descriptions.append(desc)
    
    #Get the company profile
    try:
        prof = driver.find_element_by_xpath('.//div[@class="FYwKg _3gJU3_0 _1yPon_0"]/div[2]/div[1]').text
    except:
        prof = "None"
    comp_profs.append(prof)
    
    # CLUSTER 1
    
    # For Job Levels
    try:
        # Search in first line (original place)
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Tingkat Pekerjaan":
            level.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)
        else:
            level.append("None")
        
    except:
        # Level is always placed in first line, so if the xpath is not found, then there's no job level information
        level.append("None")

    # For Qualification
    try:
        # Search in second line (original place)
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Kualifikasi":
            qualifications.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)

        # Search in first line (this part can be moved up to first line if there's no job level info)
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Kualifikasi":
            qualifications.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)

        else:
            qualifications.append("None")
            
    except:
        # if the 2 possible xpath cannot be found, then there's no qualification info 
        qualifications.append("None")

    # For job experience
    try:
        # Search in third line (original place)
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[3]/div/div/div[1]/span').text == "Pengalaman Kerja":
            experiences.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[3]/div/div/div[2]/span').text)

        # Search in second line (this part can be moved up to second line if there's no qualification info or it's moved up to first line)
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Pengalaman Kerja":
            experiences.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)

        # Search in first line (this part can be moved up to first line if there're no qualification and job level info)
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Pengalaman Kerja":
            experiences.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)    

        else:
            experiences.append("None")
            
    except:
        experiences.append("None")

    # For job type
    try:
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[4]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[4]/div/div/div[1]/span').text == "Jenis Pekerjaan":
            types.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[4]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[4]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span').text == "Jenis Pekerjaan":
            types.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[3]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Jenis Pekerjaan":
            types.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Jenis Pekerjaan":
            types.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)

        else:
            types.append("None")
            
    except:
        types.append("None")

    # For specialization
    try:
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[5]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[5]/div/div/div[1]/span').text == "Spesialisasi Pekerjaan":
            specialization.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[5]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[5]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[4]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[4]/div/div/div[1]/span').text == "Spesialisasi Pekerjaan":
            specialization.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[4]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[4]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[3]/div/div/div[1]/span').text == "Spesialisasi Pekerjaan":
            specialization.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[3]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[3]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Spesialisasi Pekerjaan":
            specialization.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)

        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Spesialisasi Pekerjaan":
            specialization.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span|//*[@id="contentContainer"]/div/div[2]/div/div[1]/div/div[3]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)

        else:
            specialization.append("None")
            
    except:
        specialization.append("None")
        
    # CLUSTER 2
    
    # For Company Scales 
    try:
        # Search in the first line
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Ukuran Perusahaan":
            company_scales.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)
        
        # Search in the second line
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Ukuran Perusahaan":
            company_scales.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)
    
        else:
            company_scales.append("None")
    
    except:
        company_scales.append("None")
        
    # For period of resume being processed
    try:
        # Search in the second line
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Waktu Proses Lamaran":
            processed.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)

        # Search in the first line
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Waktu Proses Lamaran":
            processed.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)
    
        # Search in the third line
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span').text == "Waktu Proses Lamaran":
            processed.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/div[2]/span').text)
        
        else:
            processed.append("None")
            
    except:
        processed.append("None")
        
    # For benefits
    try:
        # Search in fourth line
        if driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[4]/div/div/div[1]/span').text == "Tunjangan dan Lain-lain":
            benefits.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[4]/div/div/div[2]/span').text)
    
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/div[1]/span').text == "Tunjangan dan Lain-lain":
            benefits.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[3]/div/div/div[2]/span').text)
    
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[1]/span').text == "Tunjangan dan Lain-lain":
            benefits.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[2]/div/div/div[2]/span').text)
            
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[1]/span').text == "Tunjangan dan Lain-lain":
            benefits.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[1]/div/div/div[2]/span').text)
            
        elif driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[5]/div/div/div[1]/span').text == "Tunjangan dan Lain-lain":
            benefits.append(driver.find_element_by_xpath(r'//*[@id="contentContainer"]/div/div[2]/div/div[2]/div[2]/div[2]/div/div[2]/div/div/div[5]/div/div/div[2]/span').text)
    
        else:
            benefits.append("None")
            
    except:
        benefits.append("None")
```

```python
# Bring the information into one Dataframe

df = pd.DataFrame()

df['Title'] = titles
df['Company'] = companies
df['Location'] = locations
df['Salary'] = salaries
df['Uploaded Time'] = times
df['Job Level'] = level
df['Qualification'] = qualifications
df['Experience Years Needed'] = experiences
df['Job Type'] = types
df['Company Specialization'] = specialization
df['Company Scale'] = company_scales
df['Processed Time'] = processed
df['Benefit'] = benefits
df['Description'] = descriptions
df['Link'] = links
```
